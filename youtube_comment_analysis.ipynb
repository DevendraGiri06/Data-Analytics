#### Does Punctuations in title and tags have any relations with views

Title and tags have any relation with views, likes, dislikes comments?

```python
full_df['title']
```

```python
full_df['title'][0]
```

```python
import string
```

```python
string.punctuation
```

```python
# how many punctuation on title[0] 
[char for char in full_df['title'][0] if char in string.punctuation]
```

```python
# count the punctuation 
len([char for char in full_df['title'][0] if char in string.punctuation])
```

```python
# create a function all title 
def punc_count(text):
    return len([char for char in text if char in string.punctuation])
```

```python

full_df['title'].apply(punc_count)
```

```python
# only some sample apply 
sample = full_df[0:10000]
```

```python
sample['count_punc'] = sample['title'].apply(punc_count)
```

```python
sample['count_punc']
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='count_punc' , y='views' , data=sample)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='count_punc' , y='likes' , data=sample)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='count_punc' , y='dislikes' , data=sample)
plt.show()
```
